```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
img = cv2.VideoCapture('score.jpg')
```

```python
# Take each frame
_, frame = img.read()

# Convert BGR to HSV
#hsv = cv2.cvtColor(frame, cv2.COLOR_RGB2HSV)

# define range of blue color in HSV
lower_blue = np.array([50,215,255])
upper_blue = np.array([50,175,255])

# Threshold the HSV image to get only blue colors
mask = cv2.inRange(frame, lower_blue, upper_blue)

# Bitwise-AND mask and original image
res = cv2.bitwise_and(frame,frame, mask= mask)

plt.figure()
plt.imshow(frame)
plt.figure()
plt.imshow(mask)
plt.figure()
plt.imshow(res)
```
